```python
import pandas as pd
import json

with open("./response_times.json",'r') as fp:
    data = json.load(fp)
```

```python
import matplotlib.pyplot as plt
plt.plot(data['response_times'])
plt.title('Response Times Measured over 10,000 Random Queries')
plt.xlabel('Query Number')
plt.ylabel('Response Time (s)')
plt.savefig('response_times.png')
```
